# Data Cleaning

Our Data Cleaning was done mainly in excel
Cleaning for respective Columns



```python
MCData = pd.read_csv('Moviescombined.csv')
MCData.head()
```

```python
MCData.info()
```

```python
df1 = MCData.copy()
df1 = df1.dropna()
df1.head()
```

```python
def year_to_decade(year):
    """
    Returns the decade 
    """
    if year >= 2020:
        return '2020s'
    elif year >= 2010:
        return '2010s'
    elif year >=2000:
        return '2000s'
    elif year >=1990:
        return '1990s'
    elif year >=1980:
        return '1980s'
    elif year >=1970:
        return '1970s'
    elif year >=1960:
        return '1960s'
    elif year >=1950:
        return '1950s'
    elif year >=1940:
        return '1940s'
    elif year >=1930:
        return '1930s'
    elif year >=1920:
        return '1920s'
    elif year >=1910:
        return '1910s'
    else:
        return '1900s'



decade = []
for i in df1['Year']:
    decade.append(year_to_decade(i))

df1.insert(4, 'Decade', decade)
```

```python
df1.info()
```

```python
df1.to_csv('decade-cleaned-movie-dataset.csv')
```
